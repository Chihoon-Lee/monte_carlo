#Using Python to approximate the area of a circle (Monte Carlo Method)

The goal is to find the approximate area of a circle with radius 1 without using $\pi$. The idea is to generate a lot of points inside of a square with side length 2 and center at (0,0), and see how many of them fall inside of the circle which is inscribed within the square. 

Random x-values and y-values are chosen from the range (-1,1), to generate points on a plane (inside of the square). In order for a point (x,y) to be inside the circle which is inscribed in the square, the distance between (0,0) and (x,y) must be less than or equal to 1. Therefore if the inequality $x^2 + y^2 \leq 1$ is true, then the point (x,y) must be inside the circle (the inequality is derived from the Pythagorean Theorem). 

The number of points within the circle divided by the number of points within the square (number of points in total) will roughly be equal to the area of the circle divided by the area of the square. Therefore, the area of the circle is approximately
$$
4\times\frac{\text{number of points inside circle}}{\text{number of points in total}}
$$ which amounts to about 3.14, or $\pi$.

```python
#circle with center (0,0), radius 1
import random
numberofdots = 0
numberinsidecircle = 0
while numberofdots < 1000000:
  x = random.uniform(-1,1) #random x-coordinate
  y = random.uniform(-1,1) #random y-coordinate
  d = (x**2 + y**2) #distance from point to (0,0)
  if d <= 1: #calculating if point is inside circle or not
    numberinsidecircle = numberinsidecircle + 1
  numberofdots = numberofdots + 1
z = numberinsidecircle/numberofdots #equal to pi/4
print(z*4) #approximate estimation of pi, area of unit circle
```
