The goal is to find the approximate circumference of a circle with radius 1, without using $\pi$. The idea is to have a square circumscribed around the circle with radius 1, and to have another smaller circle with a radius of about $1-\epsilon$ inside of the bigger circle. Here, we set $\epsilon = 0.001$. This will create a thin ring, called an annulus. 

Intuitively, the area of the annulus can be approximated by splitting the annulus into, say, 100 equal parts. Each of these parts is very similar to the shape of a rectangle. Generally, it would be similar to the shape of a trapezoid if $\epsilon$ is not very small. But in our case, it will look more like a rectangle since $\epsilon$ is very small and the two circumferences of the circles are similar. When we combine each of these parts, the area will be similar to $\epsilon\times (\text{circumference of the unit circle})$.
In fact, the exact area of the annulus is $\pi(R^2 - r^2)$, where $R=1$ and $r=1-\epsilon$. From here, 
$\pi(R^2 - r^2) = \pi(R-r)(R+r)
=\pi(\epsilon)(2r-\epsilon) \approx \pi(2r)(\epsilon)$.

In order to find the approximate area of the annulus, random points are selected from inside the square the square. Then, the number of points inside the annulus divided by the nubmer of points inside the square is approximately equal to the area of the annulus divided by the area of the square. Therefore, the area of the annulus is about 
$$
4\times\frac{\text{number of points inside annulus}}{\text{number of points in total}}.
$$ To find the approximate circumference, divide the approximate area of the annulus by $\epsilon$, which is about $2\pi \approx 6.28$ .

```python
import random
import math
numberofdots = 0
annulus = 0
epsilon = .001 
while numberofdots < 50000000:
  x = random.uniform(-1,1) 
  y = random.uniform(-1,1) 
  d = math.sqrt(x**2 + y**2) 
  if 1-epsilon<d<1: #calculating if the point is inside of the annulus or not (between 1 and 0.999)
    annulus = annulus + 1
  numberofdots = numberofdots + 1
annulus_area = annulus/numberofdots*4 #equal to area of the annulus
print(annulus_area/epsilon)
```
