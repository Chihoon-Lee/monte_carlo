```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
rect1 = matplotlib.patches.Rectangle((-1,-1), 2,2, color = None)
circle1 = matplotlib.patches.Circle((0,0), radius=1, color='black')
ax.add_patch(circle1)
plt.xlim([-1,1])
plt.ylim([-1,1])
plt.show()
```
